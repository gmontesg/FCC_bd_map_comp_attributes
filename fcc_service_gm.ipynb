```python
import pandas as pd
```

```python
def speed_loc(df,speed):
    aux=df[(df['max_advertised_download_speed']>=speed[0]) & (df['max_advertised_upload_speed']>=speed[1])]['location_id'].unique()
    return aux
```

```python
infile='fcc_db_matched_ids_technology.csv'
df2=pd.read_csv(infile)
valid_tech_codes  = [10,40,50,71,0]
invalid_tech_codes = [60,61,70]
df2=df2[df2['technology_code'].isin(valid_tech_codes) & df2['low_latency']==1].reset_index(drop=True)
df2.drop(columns='low_latency',inplace=True)
df2.head()
```

```python
#speeds=[[1000,100],[200,25],[100,20],[25,3]]
speeds=[[100,20],[25,3]]
labels=['served','underserved','unserved']

dfc=pd.DataFrame(columns=['location_id','service'])

for i in range(len(speeds)):
    a=speed_loc(df2,speeds[i])
    data={'location_id':a,'service':a.shape[0]*[labels[i]]}
    aux=pd.DataFrame(data=data)
    dfc=dfc.append(aux,ignore_index=True)
    df2=df2[~df2['location_id'].isin(a)]
    
a=df2['location_id'].unique()
data={'location_id':a,'service':a.shape[0]*['unserved']}
aux=pd.DataFrame(data=data)
dfc=dfc.append(aux,ignore_index=True)
```
